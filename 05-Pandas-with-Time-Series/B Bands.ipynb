```python
import numpy as np
import pandas as pd
import matplotlib as plt
%matplotlib inline 
```

```python
df = pd.read_csv('time_data/walmart_stock.csv', index_col='Date', parse_dates=True)
```

```python
df.head()
```

```python
df['Open'].plot(figsize=(16,6))
```

```python
df['Open'].plot()
df.rolling(90).mean()['Open'].plot(figsize=(16,6))
```

```python
df['Open 30 Day MA'] = df.rolling(30).mean()['Open']
df[['Open', 'Open 30 Day MA']].plot(figsize=(16,6))
```

```python
df['Close'].expanding().mean().plot()
```

```python
# Close 20 MA
df['Close: 20 Day Mean'] = df['Close'].rolling(20).mean()

# Upper = 20MA + 2 * std(20)
df['Upper'] = df['Close: 20 Day Mean'] + 2 * (df['Close'].rolling(20).std())

# Lower = 20MA - 2 * std(20)
df['Lower'] = df['Close: 20 Day Mean'] - 2 * (df['Close'].rolling(20).std())


df[['Upper', 'Close', 'Lower']].tail(200).plot(figsize=(16,6))
```
