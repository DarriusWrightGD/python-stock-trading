```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x[0] = 0
x
```

```python
y
```

```python
#functional
plt.plot(x,y,'b-')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y, 'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# OO

fig = plt.figure()

axes = fig.add_axes([0.1,0.1, 1,1])
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,.2,.3])

axes1.plot(x,y)
axes1.set_title('Large Plot')

axes2.plot(y,x)
axes2.set_title('Small Plot')
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2)
#axes.plot(x,y)
# for current_ax in axes:
#     current_ax.plot(x,y)

plt.tight_layout()
axes[0].plot(x,y)
axes[0].set_title('First title')

axes[1].plot(y,x)
axes[1].set_title('Second title')

```

```python
# fig = plt.figure(figsize=(5,2), dpi=100)
# ax = fig.add_axes([0,0,1,1])
# ax.plot(x,y)

fig, axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2))

axes[0].plot(x,y)
axes[1].plot(x,y)

plt.tight_layout()
```

```python
fig.savefig('my_picture.svg', dpi=500)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_title('Title')
ax.set_ylabel('Y')
ax.set_xlabel('X')
ax.plot(x,x**2, label='Squared')
ax.plot(x,x**3, label='Cubed')

ax.legend()
```
